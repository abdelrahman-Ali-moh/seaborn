```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
glue=sns.load_dataset('glue')
```

```python
glue.head()
```

```python
sns.distplot(glue['Score'],kde=True,bins=10,color='g')
```

```python
sns.jointplot(x='Year',y='Score',data=glue,kind='reg')
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips,hue='smoker')
```

```python
sns.swarmplot(x='sex',y='total_bill',data=tips,split=False)
sns.violinplot(x='sex',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.stripplot(x='sex',y='total_bill',data=tips,hue='smoker',jitter=True)
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='swarm')
```

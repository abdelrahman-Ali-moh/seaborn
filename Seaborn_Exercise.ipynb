```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'],kde=False,color='red',bins=40)
```

```python
sns.boxplot(x='class',
    y='age',data=titanic,palette='rainbow',hue='class')
```

```python
sns.swarmplot(x='class',
    y='age',data=titanic,palette='Set2',hue='class')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic')
```

```python
t=sns.FacetGrid(data=titanic,col='sex')
t.map(sns.distplot,'age')
```

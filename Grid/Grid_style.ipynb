```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
sns.lmplot(x='sepal_length',y='petal_width',data=iris,hue='species',markers=['o','v','*'],scatter_kws={'s':200},palette='seismic')
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='sepal_length',y='petal_width',data=iris,col='species',aspect=0.6,size=6)
sns.despine(left=True,
    bottom=False,)
```
